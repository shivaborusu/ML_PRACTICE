```python
from sklearn.datasets import load_boston
```

```python
input_df = pd.DataFrame(data=load_boston().data, columns=load_boston().feature_names)
```

```python
input_df.head()
```

```python
target_df = pd.DataFrame(data=load_boston().target, columns=['target'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_df, target_df, test_size=0.2, random_state=42, shuffle=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(13,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop', loss='mse', metrics=["accuracy"])
```

```python
x_train.shape
```

```python
model.fit(x_train, y_train.values, batch_size=10, epochs=300)
```

```python
#returns history of loss
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df
```

```python
loss_df.plot()
```
